```python
def parse_genres(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return x  

df_movies_metadata_final = df_movies_metadata.drop(columns=['belongs_to_collection', 'homepage', 'video'])
df_movies_metadata_final['genres'] = df_movies_metadata_final['genres'].apply(parse_genres)
df_movies_metadata_final['production_companies'] = df_movies_metadata_final['production_companies'].apply(parse_genres)
df_movies_metadata_final['production_countries'] = df_movies_metadata_final['production_countries'].apply(parse_genres)
df_movies_metadata_final['spoken_languages'] = df_movies_metadata_final['spoken_languages'].apply(parse_genres)


df_movies_metadata_final
```

```python
genres = {}
def collect_genres(row):
    for i in row:
        if i.get('id') and i.get('name'):
            genres[i.get('id')] = i.get('name')

df_movies_metadata_final['genres'].apply(lambda row : collect_genres(row))
genres
```

```python
df_genres = pd.DataFrame(columns=['id_genre','genre'])
for i, name in genres.items():
    df_genres.loc[len(df_genres)] = [i, name]
df_genres.to_csv('clean/genres.csv', index=False)
```

```python
rows = []

for _, row in df_movies_metadata_final.iterrows():
    id_movie = row['id']
    for genre in row['genres']:
        if genre.get('id'):
            rows.append({'id_movie': id_movie, 'id_genre': genre['id']})

df_genres_movies = pd.DataFrame(rows)
df_genres_movies.to_csv('clean/genres_movies.csv', index=False)
```

```python
df_languages = pd.DataFrame(columns=['iso','name'])

df_movies_metadata_final['spoken_languages'] = df_movies_metadata_final['spoken_languages']
def is_nan(x):
    return isinstance(x, float) and np.isnan(x)

languages = {}

for i in df_movies_metadata_final.drop(columns=['genres', 'production_countries'])['spoken_languages']:
    if not is_nan(i) :
        if len(i) > 0:
            for language in i:
                if language.get('iso_639_1') and language.get('name'):
                    languages[language.get('iso_639_1')] = language.get('name')
for i, name in languages.items():
    df_languages.loc[len(df_languages)] = [i, name]
df_movies_metadata_final.drop(columns='genres', inplace=True)
df_languages.to_csv('clean/languages.csv', index=False)
```

```python
rows = []

for _, row in df_movies_metadata_final.iterrows():
    id_movie = row['id']
    if not is_nan(row['spoken_languages']):
        for language in row['spoken_languages']:
            if language.get('iso_639_1') and language.get('name'):
                rows.append({'id_movie': id_movie, 'iso': language.get('iso_639_1')})

df_language_movies = pd.DataFrame(rows)
df_language_movies.to_csv('clean/language_movies.csv', index=False)
df_movies_metadata_final.drop(columns='spoken_languages', inplace=True)
```

```python
df_countries = pd.DataFrame(columns=['iso','name'])

df_movies_metadata_final['production_countries'] = df_movies_metadata_final['production_countries']

countries = {}

for i in df_movies_metadata_final['production_countries']:
    if not is_nan(i) and not isinstance(i, float):
        if len(i) > 0:
            for country in i:
                if country.get('iso_3166_1') and country.get('name'):
                    countries[country.get('iso_3166_1')] = country.get('name')
for i, name in countries.items():
    df_countries.loc[len(df_countries)] = [i, name]
df_countries.to_csv('clean/countries.csv', index=False)

rows = []

for _, row in df_movies_metadata_final.iterrows():
    id_movie = row['id']
    if not is_nan(row['production_countries']) and not isinstance(row['production_countries'], float):
        for country in row['production_countries']:
            if country.get('iso_3166_1') and country.get('name'):
                rows.append({'id_movie': id_movie, 'iso': country.get('iso_3166_1')})

df_production_countries_movies = pd.DataFrame(rows)
df_production_countries_movies.to_csv('clean/production_countries_movies.csv', index=False)
df_movies_metadata_final.drop(columns='production_countries', inplace=True)
```

```python
df_companies = pd.DataFrame(columns=['id','name'])

df_movies_metadata_final['production_companies'] = df_movies_metadata_final['production_companies']

companies = {}

for i in df_movies_metadata_final['production_companies']:
    if not is_nan(i) and not isinstance(i, float) and not isinstance(i, bool)  :
        if len(i) > 0:
            for company in i:
                if company.get('id') and company.get('name'):
                    companies[company.get('id')] = company.get('name')

for id, name in companies.items():
    df_companies.loc[len(df_companies)] = [id, name]
df_companies.to_csv('clean/companies.csv', index=False)

rows = []

for _, row in df_movies_metadata_final.iterrows():
    id_movie = row['id']
    if not is_nan(row['production_companies']) and not isinstance(row['production_companies'], float) and not isinstance(row['production_companies'], bool)  :
        for company in row['production_companies']:
            if company.get('id') and company.get('name'):
                rows.append({'id_movie': id_movie, 'id': company.get('id')})

df_production_companies_movies = pd.DataFrame(rows)
df_production_companies_movies.to_csv('clean/production_companies_movies.csv', index=False)
df_movies_metadata_final.drop(columns='production_companies', inplace=True)
```
